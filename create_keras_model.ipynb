```python
import numpy as np
import random

import matplotlib.pyplot as plt
from model.tools.OneHotEncoderTools import OneHotEncoderTools
```

```python
data = np.genfromtxt("datasets/train_mnist.csv", delimiter=",")
print('reed')
```

```python
#mytest
data_shuffle = data.copy()
data_shuffle = data_shuffle[1:]
print(data_shuffle, data_shuffle.shape)
```

```python
#mytest
# data_shuffle = np.array([1,2,3,4,5])
# np.random.shuffle(data_shuffle)
# print(data_shuffle, type(data_shuffle))


# print(type(data_shuffle))
np.random.shuffle(data_shuffle)
data_shuffle = data_shuffle[:3000]
# print(data_shuffle[0])
matrix_dataset_3d = []
matrix_dataset_3d_y = []
for row in data_shuffle:
    # print(row[0])
    matrix = np.array(row[1:]).reshape((28,28, 1))/255
    encode_data = OneHotEncoderTools.to_full(int(row[0]), 10)
    matrix_dataset_3d.append(matrix)
    matrix_dataset_3d_y.append(encode_data)
```

```python
#mytest
# random.shuffle(matrix_dataset_3d)
cut_3d_x = np.array(matrix_dataset_3d)
cut_3d_y = np.array(matrix_dataset_3d_y)
print(cut_3d_y.shape)
```

```python
from keras import models
from keras import layers
import keras

model = models.Sequential()
model.add(layers.Conv2D(1, (3, 3), activation='linear', input_shape=(28, 28, 1)))
model.add(layers.Flatten())
model.add(layers.Dense(100, activation='relu'))
# model.add(layers.Dense(20, activation='sigmoid'))
model.add(layers.Dense(10, activation='softmax'))
model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.0005), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(cut_3d_x, cut_3d_y, epochs=10, batch_size=1)
```

```python
model.summary()
```

```python
import tf2onnx
```

```python
import tensorflow as tf
from tf2onnx import convert

# Загрузка модели Keras
keras_model = model  # ваша модель Keras

# Преобразование модели Keras в ONNX
onnx_model, _ = convert.from_keras(keras_model)

# Сохранение модели ONNX в файл
onnx_file_path = "saved_models/model1.onnx"
tf.io.write_graph(onnx_model, ".", onnx_file_path, as_text=False)
```
